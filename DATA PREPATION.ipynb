```python
import numpy as np
import cv2
import os
import random
import matplotlib as plt
import pickle
import matplotlib.pyplot as plt
```

```python
DIRECTORY = r"/Users/macbookpro/Desktop/dogscats/dogscats/train"
CATEGORIES = 'cats','dogs'
```

```python
# we see any one image ,its size. we break to avoid displaying all images.
```

```python
for category in CATEGORIES:
    folder = os.path.join(DIRECTORY,category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        cv2.imread(img_path)
        img_arr = cv2.imread(img_path)
        plt.imshow(img_arr)
        break
    
```

```python
# we want to resize the image using cv2
```

```python
IMG_SIZE = 227

for category in CATEGORIES:
    folder = os.path.join(DIRECTORY,category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        cv2.imread(img_path)
        img_arr = cv2.imread(img_path)
        img_arr = cv2.resize (img_arr,(IMG_SIZE,IMG_SIZE))
        plt.imshow(img_arr)
        break
```

```python
# WE NOW GIVE LABELS THE IMAGES EITHER 0'S OR 1'S by creating a list of data
```

```python
IMG_SIZE = 227
data = []
for category in CATEGORIES:
    folder = os.path.join(DIRECTORY,category)
    label= CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        cv2.imread(img_path)
        img_arr = cv2.imread(img_path)
        img_arr = cv2.resize (img_arr,(IMG_SIZE,IMG_SIZE))
        data.append([img_arr,label])
```

```python
len(data)
```

```python
# we shuffle the data
```

```python
random.shuffle(data)
```

```python
# we now create two lists one for features(cat n dogs) and the other is for lables . then we append those two lists
```

```python
X = []
Y = []

for features,labels in data:
    X.append(features)
    Y.append(labels)
    
```

```python
#we convert from list to array using numpy
```

```python
X = np.array(X)
Y = np.array(Y)
```
